```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("diamonds")
df.head(10)
```

```python
sns.barplot(x = "color",
            y = "price", data = df, palette= "spring")
```

```python
sns.displot(data = df, x = "clarity", hue = "cut", kde = True, palette = "rocket")
```

```python
sns.stripplot( data = df, x='clarity', hue='depth', palette="summer")
```

```python
sns.pairplot(data = df)
```

```python
df.plot(kind = "box", subplots = True, layout = (4, 4), sharex = False, sharey = False, figsize=(8, 12))
```

```python
df1 = df.drop("cut", axis=1)
df1 = df1.drop("color", axis=1)
df1 = df1.drop("clarity", axis=1)
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm')
```

```python
sns.countplot(x='color', data= df, palette='husl')
```

```python
sns.scatterplot(x='carat', y='price', data = df, hue='cut')
```

```python
sns.histplot(df['price'], kde=True, color='red')
```

```python
sns.boxplot(x='cut', y='price', data = df)
```

```python
sns.jointplot(x='carat', y='price', data = df, hue='cut')
```

```python
sns.pointplot(x='cut', y='price', hue='color', data = df)
```

```python
sns.lmplot(x='carat', y='price', data = df, hue='cut')
```

```python
sns.displot(df['price'], kde=True)
```
